```python
#Exercise 1
def func(time, amplitude=1, frequency=1, phase=0):
    """
    Returns a sine wave as a function of time.

    Parameters:
        time (array-like): Array of time values.
        amplitude (float): Amplitude of the sine wave. Default is 1.
        frequency (float): Frequency of the sine wave (Hz). Default is 1.
        phase (float): Phase shift (radians). Default is 0.

    Returns:
        numpy.ndarray: Sine wave values at the given time points.
    """
    return amplitude * np.sin(2 * np.pi * frequency * time + phase)
```

```python
time = np.linspace(0, 2, 1000)

# Test the function, changing parameters
y1 = func(time)
y2 = func(time, amplitude=2)
y3 = func(time, frequency=2)
y4 = func(time, phase=np.pi / 2)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(time, y1, label="A=1, f=1, φ=0")
plt.plot(time, y2, label="A=2, f=1, φ=0")
plt.plot(time, y3, label="A=1, f=2, φ=0")
plt.plot(time, y4, label="A=1, f=1, φ=π/2")
plt.title("Sine Waves with Different Parameters")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Exercise 2
G = 6.67430e-11       # gravitational constant, m^3 kg^-1 s^-2
M = 5.97219e24        # Earth's mass, kg
R = 6371000.0         # Earth's mean radius, m

def altitude_from_period(T):
    """
    Return altitude h in metres for circular orbit period T (seconds).
    """
    r = (G * M * T**2 / (4 * pi**2)) ** (1.0/3.0)
    h = r - R
    return h

def altitude_from_period(T):
    r = (G * M * T**2 / (4 * pi**2)) ** (1.0/3.0)
    return r - R

def main():
    T_str = input("Enter orbital period T in seconds : ")
    try:
        T = float(T_str)
    except ValueError:
        print("Invalid input. Please enter a number .")
        return
    h_m = altitude_from_period(T)
    h_km = h_m / 1000.0
    print(f"Altitude h = {h_km:.3f} km")

if __name__ == "__main__":
    main()
```

```python
#Exercise 3
def linfit(x, y):
    """Perform a least-squares fit to y = m*x + c and return (m, dm, c, dc)."""
    x = np.array(x)
    y = np.array(y)
    N = len(x)

    # Means
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    # Calculate slope (m) and intercept (c)
    Sxx = np.sum((x - x_mean)**2)
    Sxy = np.sum((x - x_mean)*(y - y_mean))
    m = Sxy / Sxx
    c = y_mean - m * x_mean

    # Estimate uncertainties (assuming equal weighting)
    residuals = y - (m * x + c)
    sigma_y = np.sqrt(np.sum(residuals**2) / (N - 2))
    dm = sigma_y / np.sqrt(Sxx)
    dc = sigma_y * np.sqrt(1/N + x_mean**2 / Sxx)

    return m, dm, c, dc
```

```python
# Example test data
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2.1, 2.9, 4.2, 4.9, 6.1, 6.2])

# Fit the data
m, dm, c, dc = linfit(x, y)
print(f"Slope (m) = {m:.4f} ± {dm:.4f}")
print(f"Intercept (c) = {c:.4f} ± {dc:.4f}")

# Plot data and fitted line
x_fit = np.linspace(min(x), max(x), 100)
y_fit = m * x_fit + c

plt.figure(figsize=(6,4))
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(x_fit, y_fit, color='red', label=f'Fit: y = {m:.2f}x + {c:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Least Squares Fit Example')
plt.legend()
plt.grid(True)
plt.show()
```
